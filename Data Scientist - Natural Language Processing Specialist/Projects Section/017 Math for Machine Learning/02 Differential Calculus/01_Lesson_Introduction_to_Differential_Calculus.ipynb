````python
# our change in x value
dx = 0.01
def f(x):
    return sin(x) 
sin_x = [x for x in np.arange(0,20,dx)] # 0 to 20 in increments of 0.01
sin_y = [f(x) for x in np.arange(0,20,dx)] # sin(x) for each x in 0 to 20 in increments of 0.01
print(sin_y)

plt.plot(sin_x, sin_y)
#plt.plot(sin_x, sin_deriv)
plt.show()
````

````python
# Task 1
sin_deriv = np.gradient(sin_y, dx)
print(sin_deriv)

# Task 2
plt.plot(sin_x, sin_deriv)
plt.show()
````

````python
# Task 3
dx = 1
def f(x):
    return sin(x) 
sin_x = [x for x in np.arange(0,20,dx)] # 0 to 20 in increments of 0.01
sin_y = [f(x) for x in np.arange(0,20,dx)] # sin(x) for each x in 0 to 20 in increments of 0.01
sin_deriv = np.gradient(sin_y, dx)

plt.plot(sin_x, sin_y)
plt.plot(sin_x, sin_deriv)
plt.show()
````
