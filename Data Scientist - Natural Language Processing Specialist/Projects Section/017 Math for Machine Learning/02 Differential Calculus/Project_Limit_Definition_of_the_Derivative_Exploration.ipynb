````python
import numpy as np
from math import sin, cos, log, pi
import matplotlib.pyplot as plt

def limit_derivative(f, x, h):
    """
    f: function to be differentiated 
    x: the point at which to differentiate f 
    h: distance between the points to be evaluated
    """
    # compute the derivative at x with limit definition
    return (f(x+h) - f(x)) / h


# f1(x) = sin(x)
def f1(x):
    return np.sin(x)

# f2(x) = x^4
def f2(x):
    return np.pow(x, 4)

# f3(x) = x^2*log(x)
def f3(x):
    return np.pow(x, 2) * np.log(x)

# Calculate derivatives here
print("Calculating Limit Derivative for f3")
print(limit_derivative(f3, 1, 2))
print(limit_derivative(f3, 1, 0.1))
print(limit_derivative(f3, 1, 0.00001), end="\n\n")

print("Calculating Limit Derivative for f2")
print(limit_derivative(f2, 1, 2))
print(limit_derivative(f2, 1, 0.1))
print(limit_derivative(f2, 1, 0.00001), end="\n\n")

print("Calculating Limit Derivative for f1")
print(limit_derivative(f1, 1, 2))
print(limit_derivative(f1, 1, 0.1))
print(limit_derivative(f1, 1, 0.00001), end="\n\n")


# Graph the true derivative
x_vals = np.linspace(1, 10, 200)
y_vals_f1 = [cos(val) for val in x_vals]
y_vals_f2 = [4*pow(val, 3) for val in x_vals]
y_vals_f3 = [val+2*log((val)) for val in x_vals]

# plot different approximated derivatives of f using limit definition of derivative
def plot_approx_deriv(f):
    x_vals = np.linspace(1, 10, 200)
    h_vals = [10, 1, .25, .01]

    for h in h_vals:
        derivative_values = []
        for x in x_vals:
            derivative_values.append(limit_derivative(f, x, h))
        plt.plot(x_vals, derivative_values, linestyle='--', label="h=" + str(h) )
    plt.legend()
    plt.title("Convergence to Derivative by varying h")


# plot different approximated derivatives of f using numpy gradient
def plot_approx_deriv_np_gradient(f):
    n = [5, 20, 50, 200, 1000]
    x_valss = [np.linspace(1, 10, i) for i in n]
    
    for x_vals in x_valss:
        derivative_values = np.gradient(f(x_vals), x_vals)
        plt.plot(x_vals, derivative_values, linestyle='--', label="h=" + str(np.round(np.diff(x_vals).mean(), 2)))
    plt.legend()


# Plot here
plt.figure(figsize=(30, 10))
plt.subplot(2, 3, 1)
plt.plot(x_vals, y_vals_f1, label="True Derivative", linewidth=4)
plot_approx_deriv(f1)
plt.title("Derivative of sin(x)")

plt.subplot(2, 3, 2)
plt.plot(x_vals, y_vals_f2, label="True Derivative", linewidth=4)
plot_approx_deriv(f2)
plt.title("Derivative of x^4")

plt.subplot(2, 3, 3)
plt.plot(x_vals, y_vals_f3, label="True Derivative", linewidth=4)
plot_approx_deriv(f3)
plt.title("Derivative of x^2*log(x)")

plt.subplot(2, 3, 4)
plt.plot(x_vals, y_vals_f1, label="True Derivative", linewidth=4)
plot_approx_deriv_np_gradient(f1)
plt.title("Derivative of sin(x) using numpy gradient")

plt.subplot(2, 3, 5)
plt.plot(x_vals, y_vals_f2, label="True Derivative", linewidth=4)
plot_approx_deriv_np_gradient(f2)
plt.title("Derivative of x^4 using numpy gradient")

plt.subplot(2, 3, 6)
plt.plot(x_vals, y_vals_f3, label="True Derivative", linewidth=4)
plot_approx_deriv_np_gradient(f3)
plt.title("Derivative of x^2*log(x) using numpy gradient")

plt.show()
````
