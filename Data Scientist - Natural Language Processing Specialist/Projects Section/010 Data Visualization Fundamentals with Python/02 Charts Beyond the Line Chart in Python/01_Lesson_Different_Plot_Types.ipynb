## 5. Stacked Bars

````python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
````

**Task 1**:  
- You just made a chart with two sets of sales data plotted side by side. 
- Letâ€™s instead make a stacked bar chart by using the keyword `bottom`. 
- Put the `sales1` bars on the bottom and set the `sales2` bars to start where the `sales1` bars end.

<br>

**Task 2**:  
- We should add a legend to make sure we know which set of bars corresponds to which location.
-  Label the bottom set of bars as â€œLocation 1â€ and the top set of bars as â€œLocation 2â€ and add a legend to the chart.

````python
# Task 1
plt.bar(range(len(drinks)), sales1)
plt.bar(range(len(drinks)), sales2, bottom=sales1)

# Task 2
plt.legend(["Location 1", "Location 2"])

plt.show()
````

## 7. Fill Between

````python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
````

**Task 1**:  
- We have provided a set of data representing MatplotSipâ€™s projected revenue per month for the next year in the variable `revenue`. 
- Letâ€™s plot these revenues against `months` as a line.

<br>

**Task 2**:  
- Make an axis object, store it in the variable `ax`, and then use it to set the x-ticks to `months` and the x-axis tick labels to be the months of the year, given to you in the variable `month_names`.

<br>

**Task 3**:  
- This data is a projection of future revenue. 
- We donâ€™t *know* that this will be the revenue, but itâ€™s an estimate based on the patterns of past years. 
- We can say that the real revenue will probably be plus or minus 10% of each value. 
- Create a list containing the lower bound of the expected revenue for each month, and call it `y_lower`.
- Remember that 10% less than a number would be either:
    ```py
    i - 0.1 * i
    ```
- or
    ```py
    i * 0.9
    ```
- You can use either of these in your list comprehension.

<br>

**Task 4**:  
- Create a list containing the upper bound of the expected revenue for each month, and call it `y_upper`.
- Remember that 10% more than a number would be either:
    ```py
    i + 0.1 * i
    ```
- or
    ```py
    1.1 * i
    ```
- You can use either of these in your list comprehension.

<br>

**Task 5**:  
- Use `.fill_between()` to shade the error above and below the line weâ€™ve plotted, with an alpha of `0.2`.

````python
# Task 1
plt.plot(months, revenue)

# Task 2
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)

# Task 3
# or just use numpy to multiply by 0.9, because numpy can multiply a list by a number
y_lower = [0.9*i for i in revenue]  

# Task 4
y_upper = [1.1*i for i in revenue]

# Task 5
plt.fill_between(months, y_lower, y_upper, alpha=0.2)

plt.show()
````

## 10. Histogram

````python
import csv


def convert_time_to_num(time):
    mins = int(time[-2:])
    frac_of_hour = mins/60.0
    hour = int(time[:-3])
    time = hour + frac_of_hour
    return time

sales_times_raw = []
with open('sales_times.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        sales_times_raw.append(row[2])
    sales_times_raw = sales_times_raw[1:]

sales_times = []
for time in sales_times_raw:
    sales_times.append(convert_time_to_num(time))
````

**Task 1**:  
- Weâ€™ve provided data in the file **sales_times.csv** and loaded it into a list called `sales_times`. 
- This set represents the 270 sales at MatplotSipâ€™s first location from 8am to 10pm on a certain day.

<br>

**Task 2**:
- Make a histogram out of this data using the `plt.hist` function.

<br>

**Task 3**:  
- Use the `bins` keyword to create 20 bins instead of the default 10.

````python
plt.hist(sales_times, bins=20)
plt.show()
````

## 11. Multiple Histograms

````python
import csv

def convert_time_to_num(time):
    mins = int(time[-2:])
    frac_of_hour = mins/60.0
    hour = int(time[:-3])
    time = hour + frac_of_hour
    return time

sales_times_raw = []
with open('sales_times.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        sales_times_raw.append(row[2])
    sales_times_raw = sales_times_raw[1:]

sales_times1 = []
for time in sales_times_raw:
    sales_times1.append(convert_time_to_num(time))

sales_times2_raw = []
with open('sales_times_s2.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        sales_times2_raw.append(row[2])
    sales_times2_raw = sales_times2_raw[1:]

sales_times2 = []
for time in sales_times2_raw:
    sales_times2.append(convert_time_to_num(time))


plt.hist(sales_times1, bins=20)
plt.show()
````

**Task 1**:  
- Weâ€™ve provided another dataset in the file **sales_times_s2.csv** that represents the 371 sales at MatplotSipâ€™s first location from 8am to 10pm on the same day. 
- This data has the same structure as the sales times data from store 1, with an `id`, a `card_no`, and a `time`. 
- Take a look at the data in the csv and familiarize yourself with it.
- Weâ€™ve imported the times into a list called `sales_times2`. 
- Youâ€™ll only be interacting with the lists `sales_times1` and `sales_times2`

<br>

**Task 2**:  
- Plot the histogram of times from the second location on top of the one from the last exercise.

<br>

**Task 3**:  
- Notice that the histogram we plotted second completely obscures the first histogram we plotted.
-  Modify the transparency value of both histograms to be `0.4` so that we can see the separate histograms better.

<br>

**Task 4**:  
- Normalize both the histograms so that we can compare the patterns between them despite the differences in sample size.

````python
plt.hist(sales_times1, bins=20, alpha=0.4, density=True)
plt.hist(sales_times2, bins=20, alpha=0.4, density=True)

plt.show()
````
