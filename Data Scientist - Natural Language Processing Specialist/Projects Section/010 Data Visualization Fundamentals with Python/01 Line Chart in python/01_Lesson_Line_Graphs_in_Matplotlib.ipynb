## 5. Axis and Labels

## 6. Labeling the Axes

```python
x = range(12)
y = [3000, 3005, 3010, 2900, 2950, 3050, 3000, 3100, 2980, 2980, 2920, 3010]
plt.plot(x, y)
plt.axis([0, 12, 2900, 3100])

plt.show()
```

- Label the x-axis `'Time'`.
- Label the y-axis `'Dollars spent on coffee'`.
- Add the title `'My Last Twelve Years of Coffee Drinking'`.

```python
plt.plot(x, y)
plt.axis([0, 12, 2900, 3100])
plt.xlabel('Time')
plt.ylabel('Dollars spent on coffee')
plt.title('My Last Twelve Years of Coffee Drinking')
plt.show()
```

## 7. Subplots

```python
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]
```

**Task 1**  
- We have defined the lists `months`, `temperature`, and `flights_to_hawaii` for you.
- Using the `plt.subplot` command, plot `temperature` vs `months` in the left box of a figure that has 1 row with 2 columns.

<br>

**Task 2**  
- Plot `flights_to_hawaii` vs `temperature` in the same figure, to the right of your first plot. 
- Add the parameter `"o"` to the end of your call to `plt.plot` to make the plot into a scatterplot, if you want!

```python
# Task 1
plt.subplot(1, 2, 1)
plt.plot(months, temperature)
plt.title('Temperature over the months')
plt.xlabel('Month')
plt.ylabel('Temperature')

# Task 2
plt.subplot(1, 2, 2)
plt.plot(temperature, flights_to_hawaii, "o")
plt.title('Temperature vs. number of flights')
plt.xlabel('Temperature')
plt.ylabel('Flights to Hawaii')

plt.subplots_adjust(wspace=0.5) # Adjust the space between the two plots
plt.show()
```

## 10. Modify Ticks

```python
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]

months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]

plt.xlabel("Months")
plt.ylabel("Conversion")

plt.plot(months, conversion)

plt.show()
```

**Task 1**  
- Letâ€™s imagine we are working for a company called Dinnersaur, that delivers dinners to people who donâ€™t want to cook. 
- Dinnersaur recently started a new service that is subscription-based, where users sign up for a whole year of meals, instead of buying meals on-demand.
- We have plotted a line for you in the editor representing the proportion of users who have switched over to this new service since it was rolled out in January. 
- First, save the set of axes in a variable called `ax`. 
- We will use `ax` to set the x- and y-ticks and labels to make this graph easier to read.

<br>

**Task 2**  
- Using ax, set the x-ticks to be the `months` list.

<br>

**Task 3**  
- Set the x-tick labels to be the `month_names` list.

<br>

**Task 4**  
- Set the y-ticks to be `[0.10, 0.25, 0.5, 0.75]`.

<br>

**Task 5**  
- Label the y-ticks to be the percentages that correspond to the values `[0.10, 0.25, 0.5, 0.75]`, instead of decimals.

```python
plt.xlabel("Months")
plt.ylabel("Conversion")

plt.plot(months, conversion)

# Task 1
ax = plt.subplot()

# Task 2
ax.set_xticks(months)

# Task 3
ax.set_xticklabels(month_names)

# Task 4
ax.set_yticks([0.10, 0.25, 0.5, 0.75])

# Task 5
ax.set_yticklabels(['10%', '25%', '50%', '75%'])

plt.show()
```

```python
import random

# Task 1
x = range(20)
y1 = [random.randint(10,20) for _ in x]
y2 = [y + random.randint(-10,10) for y in y1]

if len(x) != len(y1):
    raise ValueError("x and y must have the same number of elements")

plt.plot(x, y1, color="pink", marker="o")
plt.plot(x, y2, color="gray", marker="o")
plt.title("Two Lines on One Graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")
plt.legend(["y1", "y2"], loc=4)
plt.show()
```

```python
ax = plt.subplot(1, 2, 1)
ax.plot(x, y1, color="pink", marker="o")
ax.set_xticks(range(0, 21, 2))


ax2 = plt.subplot(1, 2, 2)
ax2.plot(x, y2, color="gray", marker="o")
ax2.set_xticks(range(0, 21, 2))

plt.show()
```
