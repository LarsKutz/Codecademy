```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from chart_studio.plotly import plot, iplot
```

## 3. Supervised Learning: Regression

**Task 1**  
- Run the code to train the model and predict the monthly rent for a 950-square-foot apartment in a neighborhood with 2 burglaries in the last year.

<br>

**Task 2**  
- Change the variables `square_footage` and `number_of_burglaries` to different values.

```python
# Load the data
housing_data = pd.read_csv('housing_data.csv')
X = housing_data[['Sq ft', 'Burglaries']]
y = housing_data['Rent']

# Create the model
reg = LinearRegression()

# Train the model
reg.fit(X.values, y.values)

square_footage_1 = 1250
number_of_burglaries_1 = 2

square_footage_2 = 950
number_of_burglaries_2 = 7

y_pred_1 = reg.predict(np.array([square_footage_1, number_of_burglaries_1]).reshape(1, 2))
y_pred_2 = reg.predict(np.array([square_footage_2, number_of_burglaries_2]).reshape(1, 2))


y_pred_1, y_pred_2
```

## 4. Supervised Learning: Classification

**Task 1**  
- Run the code to train the model and see whether someone who identifies a picture in 5 seconds can pass as a human.

<br>

**Task 2**  
- Change the variable `time_to_identify_picture` to a different value. 
- Run the code again. 
- Does the model still think that youâ€™re a human?
- If you want to experiment more you can change the value again and see how that affects the outcome.

```python
# Load the data
photo_id_times = pd.read_csv('photo_id_times.csv')

# Separate the data into independent and dependent variables
X = np.array(photo_id_times['Time to id photo']).reshape(-1, 1)
y = photo_id_times['Class']

# Create a model and fit it to the data
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)

time_to_identify_picture_1 = 10
time_to_identify_picture_2 = 1

# Make a prediction based on how long it takes to identify a picture
y_pred_1 = neigh.predict(np.array(time_to_identify_picture_1).reshape(1, -1))
y_pred_2 = neigh.predict(np.array(time_to_identify_picture_2).reshape(1, -1))

print("Prediction 1:")
if y_pred_1 == 1:
    print("We think you're a robot.")
else:
    print("Welcome, human!")

print("\nPrediction 2:")
if y_pred_2 == 1:
    print("We think you're a robot.")
else:
    print("Welcome, human!")
```

## Unsupervised Learning

**Task 1**  
- Press run to see the clusters!

```python
# Load the data
media_usage = pd.read_csv('media_usage.csv')

# Create the model
kmeans = KMeans(n_clusters=3)

# Fit the model to the data
kmeans.fit(media_usage)

labels = kmeans.predict(media_usage)

# Plot the data in a 3D plot 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(media_usage['Hours reading text'], media_usage['Hours watching video'], media_usage['Hours in VR'], c=labels)
ax.set_xlabel('Hours reading text')
ax.set_ylabel('Hours watching video')
ax.set_zlabel('Hours in VR')
plt.show()
```
